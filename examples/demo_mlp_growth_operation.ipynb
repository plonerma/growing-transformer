```python
import torch
import numpy as np
from sandbox import SimpleModel, SineToyDataset
from torch.utils.data import DataLoader
import copy
import matplotlib.pyplot as plt
import random
```

## Simple Model

```python
torch.manual_seed(0)

# keep same initialization
original_model = SimpleModel(1, 1, 4, 6, 6)
```

```python
model = original_model
optim = torch.optim.SGD(model.parameters(), lr=0.1)
criterion = torch.nn.MSELoss()
losses = list()
batch_size = 32
train_data = SineToyDataset(8000)

for epoch in range(10):
    loss_sum = 0
    
    batch_loader = DataLoader(train_data, batch_size=batch_size, shuffle=epoch > 0)
    
    for train_x, train_y in batch_loader:

        optim.zero_grad()
        y = model(train_x)
        loss = criterion(y, train_y)
        loss_sum += loss.data
        loss.backward()
        optim.step()
        
    losses.append(loss_sum / len(batch_loader))


print(f"loss: {loss.data} - size: {sum(p.numel() for p in model.parameters())}")

plt.subplots(figsize=(10, 4))
plt.plot(losses)
plt.gca().xaxis.grid(True)
plt.show()
```

```python
plt.subplots(figsize=(12, 8))

model.eval()

batch_loader = DataLoader(train_data, batch_size=batch_size, shuffle=False)

for train_x, train_y in batch_loader: break


x = torch.linspace(-5, 5, 100)[:, None]



y = model(x).detach().numpy()
x = x.detach().numpy()
plt.plot(x, y, c='tab:orange')

plt.scatter(train_x.detach(), train_y.detach(), marker='+')
plt.ylim([-1, 1])
plt.show()
```

# Growing the model

## Influence of new neuron (with different inits)

```python

cmap = plt.get_cmap("viridis")
vmax = 4


torch.manual_seed(0)


plt.subplots(figsize=(12, 8))
for step in np.linspace(0, vmax, 20)[::-1]:
    x = torch.linspace(-5, 5, 100)[:, None]
    
    model = copy.deepcopy(original_model)
    
    if step is not None:
        model.config.update(dict(num_novel=4, step_size=step, split=False))
        for m in model.growing_modules():
            m.grow()

    y = model(x).detach().numpy()
    x = x.detach().numpy()
    plt.plot(x, y, c=cmap(1-step/vmax))

plt.scatter(train_x.detach(), train_y.detach(), marker='x')
plt.ylim([-1, 1])
plt.show()
```

## Influence of neuron split noise

```python
torch.manual_seed(0)
vmax = 4

plt.subplots(figsize=(12, 8))

for step in np.linspace(0, vmax, 20)[::-1]:
    x = torch.linspace(-5, 5, 100)[:, None]
    
    model = copy.deepcopy(original_model)
    if step is not None:
        model.config.update(dict(split=True, num_novel=0, step_size=step))
        for m in model.growing_modules():
            m.grow()
        
        
    y = model(x).detach().numpy()
    x = x.detach().numpy()
    plt.plot(x, y, c=cmap(1-step/vmax))

plt.scatter(train_x.detach(), train_y.detach(), marker='x')
plt.ylim([-1, 1])
plt.show()
```
